# **Refatoração dos dados**

```python
import pandas as pd  # Importação da biblioteca pandas
import matplotlib.pyplot as plt 
# Leitura do arquivo CSV
data = pd.read_csv('googleplaystore.csv')

data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')

data['App'] = data['App'].str.lower()
# Remover duplicatas mantendo o maior valor de 'Reviews'
data_sem_duplicadas = data.sort_values('Reviews', ascending=False).drop_duplicates(subset=['App'], keep='first')

# Remoção de caracteres não numéricos da coluna 'Installs'
data_sem_duplicadas['Installs'] = data_sem_duplicadas['Installs'].str.replace(r'\D', '', regex=True)

# Conversão para int, convertendo valores inválidos em NaN e removendo esses valores
data_sem_duplicadas['Installs'] = pd.to_numeric(data_sem_duplicadas['Installs'], errors='coerce')

# Conversão final da coluna 'Installs' para int
data_sem_duplicadas['Installs'] = data_sem_duplicadas['Installs'].astype(int)
data_sem_duplicadas['Price'] = data_sem_duplicadas['Price'].str.replace('$', '').astype(float)
data_sem_duplicadas['Category'] = data_sem_duplicadas['Category'].str.replace('_', ' ')
```

# **Ordenação do DataFrame pela coluna 'Installs' em ordem decrescente**
#### Critério de desenpate foi o número de reviews

```python
top_5_apps = data_sem_duplicadas.sort_values(['Installs','Reviews'], ascending=False, inplace=False).head(5)

display(top_5_apps)
```

```python
fig, ax = plt.subplots(figsize=(6, 8))

ax.set_facecolor('#232321')  

plt.bar(top_5_apps['App'], top_5_apps['Installs'], color='#6DD08E')
plt.xlabel('Aplicativos', fontsize=12)
plt.ylabel('Instalações (Bilhões)', fontsize=12)
plt.title('Top 5 Aplicativos Mais Instalados na Google Play Store\n Obs.: Critério de desempate adotado foi por numeros de reviews', fontsize=16)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()

# Exibição do gráfico
plt.show()
```

```python
frequencia_aplicativos = data_sem_duplicadas['Category'].value_counts()

explode = []
for count in frequencia_aplicativos:
    if count > 300:
        explode.append(0.01)
    elif count >50:
        explode.append(0.2)  
    else:
        explode.append(0.6)


plt.figure(figsize=(16, 12))
plt.pie(
    frequencia_aplicativos,
    explode=explode,
    labels=frequencia_aplicativos.index,
    labeldistance= 1.07,
    autopct='%1.1f%%',
    pctdistance=0.94,
    startangle=27,
)

plt.legend( frequencia_aplicativos.index, title="Categorias", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))


plt.title('Frequência de Aplicativos por Categoria',fontsize=16)
plt.axis('equal')  # Para assegurar que o gráfico é desenhado como um círculo.

# Exibir o gráfico
plt.show()
```

```python
data_sem_duplicadas = data_sem_duplicadas.sort_values(['Price'], ascending=False, inplace=False)

app_maior_preco = data_sem_duplicadas.loc[data_sem_duplicadas['Price'].idxmax()]

# Configurar a figura
fig, ax = plt.subplots(figsize=(6, 4))

# Remover os eixos
ax.axis('off')

# Adicionar título
plt.title('Aplicativo mais caro', fontsize=16, weight='bold')

# Adicionar informações do app com maior preço
texto = (
    f"App mais caro\n\n"
    f"App: {app_maior_preco['App']}\n"
    f"Categoria: {app_maior_preco['Category']}\n"
    f"Preço: ${app_maior_preco['Price']:.2f}"
)


# Adicionar texto ao card
plt.text(0.5, 0.5, texto, fontsize=16, color='#6DD08E', va='center', ha='center', bbox=dict(
    facecolor='#232323',
    edgecolor='#6DD08E',
    boxstyle='round, pad=2',
    alpha=0.9,
))


plt.show()
```

```python
mature_apps = data_sem_duplicadas[data_sem_duplicadas['Content Rating'] == 'Mature 17+']

# Contar a quantidade de aplicativos com classificação 'Mature 17+'
count_mature_apps = mature_apps['Content Rating'].value_counts().get('Mature 17+', 0)

# Configurar a figura
fig, ax = plt.subplots(figsize=(6, 4))

# Remover os eixos
ax.axis('off')

# Adicionar título
plt.title('Contagem de Apps com Classificação Mature 17+', fontsize=16, weight='bold')

# Adicionar informações da contagem dos apps com classificação 'Mature 17+'
texto = (
    f"Apps classicados como Mature 17+ \nTotal:  {count_mature_apps}"
)

# Adicionar texto ao card com um estilo mais atraente
# Adicionar texto ao card
plt.text(0.5, 0.5, texto, fontsize=16, color='#6DD08E', va='center', ha='center', bbox=dict(
    facecolor='#232323',
    edgecolor='#6DD08E',
    boxstyle='round, pad=2',
    alpha=0.9,
))
# Exibir o gráfico
plt.show()
```

# **Top 10 apps mais visualizados**

```python
df = data_sem_duplicadas.sort_values(['Reviews'], ascending=False, inplace=False)

# Selecionar os top 10 aplicativos
top_10_apps = df.head(10)

# Criar um DataFrame apenas com as colunas 'App' e 'Reviews' e adicionar uma coluna de ranking
top_10_apps_ranking = top_10_apps[['App', 'Reviews']].reset_index(drop=True)

# Criar a coluna de ranking com 1°, 2°, 3°, etc.
ranking = [f"{i}°" for i in range(1, 11)]

# Adicionar a coluna de ranking ao DataFrame
top_10_apps_ranking.insert(0, 'Ranking', ranking)

# Estilizar o DataFrame para centralizar o conteúdo
styled_top_10 = top_10_apps_ranking.style.set_table_styles(
    [{'selector': 'td', 'props': [('text-align', 'left')]},
     {'selector': 'th', 'props': [('text-align', 'left')]}]
).set_properties(**{'text-align': 'left'}).hide(axis='index')

# Exibir o DataFrame estilizado
display(styled_top_10)
```

# **Top 5 Apps pagos com mais visualizações**

```python
filtered_df = data_sem_duplicadas.loc[df['Type'] == 'Paid']
top_3_apps = filtered_df.sort_values(['Reviews'], ascending=False, inplace=False)
top_3_apps = top_3_apps[['App', 'Reviews', 'Category']].reset_index(drop=True)

top_3_apps = top_3_apps.head(5)
# Criar a coluna de ranking com 1°, 2°, 3°, etc.
ranking = [f"{i}°" for i in range(1, 6)]

# Adicionar a coluna de ranking ao DataFrame
top_3_apps.insert(0, 'Ranking', ranking)

# Estilizar o DataFrame para centralizar o conteúdo
styled_top_10 = top_3_apps.style.set_table_styles(
    [{'selector': 'td', 'props': [('text-align', 'left')]},
     {'selector': 'th', 'props': [('text-align', 'left')]}]
).set_properties(**{'text-align': 'left'}).hide(axis='index')

# Exibir o DataFrame estilizado
display(styled_top_10)
```

# **App pago com melhor avaliação e o mais instalado**

```python

# Filtrar os dados onde a coluna 'Type' é igual a 'Paid'
filtered_df = data_sem_duplicadas.loc[data_sem_duplicadas['Type'] == 'Paid']

# Ordenar pelos valores das colunas 'Rating' e 'Installs' em ordem decrescente
filtered_df = filtered_df.sort_values(['Rating', 'Installs'], ascending=False, inplace=False)
# Selecionar apenas as colunas 'App', 'Installs' e 'Rating' e redefinir o índice
filtered_df = filtered_df[['App', 'Installs', 'Rating']].reset_index(drop=True).head(1)

# Selecionar o top 1 aplicativo

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(4,4))

# Remover os eixos
ax.axis('off')

# Adicionar título
plt.title('Aplicativo pago com melhor avaliação e o mais instalado', fontsize=16)

# Adicionar informações do app com maior preço
texto = (
    f"App pago melhor avaliado e o mais instalado\n\n"
    f"App: {filtered_df.loc[0, 'App']}\n"
    f"Avaliação: {filtered_df.loc[0, 'Rating']}\n"
    f"Downloads: {filtered_df.loc[0, 'Installs']:,}"
)

# Adicionar texto ao card
plt.text(0.5, 0.5, texto, fontsize=16, color='#6DD08E', va='center', ha='center', bbox=dict(
    facecolor='#232323',
    edgecolor='#6DD08E',
    boxstyle='round, pad=2',
    alpha=0.9,
))

# Mostrar o gráfico
plt.show()

```

# **Quantidade de Apps por categoria**

```python
fig, ax = plt.subplots(figsize=(20, 8))
frequencia_aplicativos = data_sem_duplicadas['Category'].value_counts()
# Definir a cor de fundo para a figura e para o eixo
##fig.patch.set_facecolor('#232323')  # Cor de fundo para a figura
ax.set_facecolor('#232321')  
plt.stem(frequencia_aplicativos.index, frequencia_aplicativos, linefmt='#6DD08E', markerfmt='D')

# Adicionar anotações para cada ponto no gráfico
posicao = 0
for category, count in frequencia_aplicativos.items():
    plt.text(posicao - 0.3,int(count)+70, int(count), fontsize=12, ha='left', va='top', color='#6DD08E')  
    posicao+=1

plt.xticks(rotation=40, ha='right')
# Adicionar títulos e rótulos
plt.title('Quantidade de Aplicativos por Categoria', fontsize=20)
plt.xlabel('Categorias', fontsize=16)
plt.ylabel('Quantidade de aplicativos', fontsize=16)


# Mostrar o gráfico
plt.show()
```

# **Grafico dispersão: Quantidade de preços de apps pagos**

```python
apps_pagos = data_sem_duplicadas.loc[data_sem_duplicadas['Type'] == 'Paid']

apps_pagos = apps_pagos['Price'].value_counts()


tamanho = []
for count in apps_pagos:
        tamanho.append(count+100)
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_facecolor('#232321')  

# Adicionar títulos e rótulos
plt.title('Frequência de preços de apps pagos', fontsize=20)
plt.xlabel('Preço ($)', fontsize=13)
plt.ylabel('Quantidade', fontsize=13)
plt.scatter(apps_pagos.index, apps_pagos, s=tamanho , color='#6DD08E')
plt.show()
```
